```python
import MetaTrader5 as mt5
import pandas as pd
from datetime import datetime
from getCandle import Rates
from plotting import CandlePlot
```

```python
rates = Rates("EURUSD",1000,mt5.TIMEFRAME_H1)
df = rates.get_rates_from_now()
df
```

```python
df_ma = df[['time', 'open', 'high', 'low', 'close']].copy()
```

```python
df_ma
```

```python
MA_LIST = [10, 20, 50, 100, 200]
for ma in MA_LIST:    
    df_ma[f'MA_{ma}'] = df_ma.close.rolling(window=ma).mean()
df_ma.dropna(inplace=True)
df_ma.reset_index(inplace=True, drop=True)
```

```python
cp = CandlePlot(df_ma)
traces = [ f"MA_{x}" for x in MA_LIST ]
cp.show_plot(line_traces=traces)
```

```python
MA_S = "MA_10"
MA_L = "MA_20"
BUY = 1
SELL = -1
NONE = 0
```

```python
df_an = df_ma[['time', 'open', 'high', 'low', 'close', MA_S, MA_L]].copy()
```

```python
df_an
```

```python
df_an['DELTA'] = df_an[MA_S] - df_an[MA_L]
df_an['DELTA_PREV'] = df_an.DELTA.shift(1)
```

```python
def is_trade(row):
    if row.DELTA >= 0  and row.DELTA_PREV < 0:
        return BUY
    elif row.DELTA < 0  and row.DELTA_PREV >= 0:
        return SELL
    return NONE
```

```python
df_an['TRADE'] = df_an.apply(is_trade, axis=1)
```

```python
df_trades = df_an[df_an.TRADE != NONE].copy()
```

```python
df_trades
```

```python
cp = CandlePlot(df_an.iloc[:80])
cp.show_plot(line_traces=[MA_S, MA_L])
```

```python
df_trades['DIFF'] = df_trades.close.diff().shift(-1)
df_trades.fillna(0, inplace=True)
```

```python
PIP_VALUE = 0.0001
```

```python
df_trades['GAIN'] = df_trades['DIFF'] / PIP_VALUE
df_trades['GAIN'] = df_trades['GAIN'] * df_trades['TRADE'] 
```

```python
df_trades.GAIN.sum()
```

```python
df_trades['GAIN_C'] = df_trades['GAIN'].cumsum()
```

```python
cp = CandlePlot(df_trades, candles=False)
cp.show_plot(line_traces=['GAIN_C'])
```

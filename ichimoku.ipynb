```python
cp = CandlePlot(df_an, candles=True)
# cp.show_plot(sec_traces=[])

cp.fig.add_trace(go.Scatter(
    x=cp.df_plot.sTime,
    y=cp.df_plot['Tenkan_sen'],
    line=dict(width=2,color="#FF0000"),
#     line_shape="spline",
    name='Tenkan_sen'
    ) )
cp.fig.add_trace(go.Scatter(
    x=cp.df_plot.sTime,
    y=cp.df_plot['Kijun_sen'],
    line=dict(width=2,color="#0000FF"),
    name='Kijun_sen'
    ) )

cp.fig.add_trace(go.Scatter(
    x=cp.df_plot.sTime,
    y=cp.df_plot['Senkou_Span_A'],
    line=dict(width=2,color="#00FFFF"),
    name='Senkou_Span_A'
    ) )

cp.fig.add_trace(go.Scatter(
    x=cp.df_plot.sTime,
    y=cp.df_plot['Senkou_Span_B'],
    line=dict(width=2,color="#00FFFF"),
    name='Senkou_Span_B'
    ) )

MARKER = '#0066FF'

df_temp = cp.df_plot[cp.df_plot.Senkou_Span_A_touch==-1]
cp.fig.add_trace(go.Candlestick(
                x=df_temp.sTime,
                open=df_temp.open,
                high=df_temp.high,
                low=df_temp.low,
                close=df_temp.close,
                line=dict(width=1), opacity=1,
                increasing_fillcolor=MARKER,
                decreasing_fillcolor=MARKER,
                increasing_line_color=MARKER,  
                decreasing_line_color=MARKER
            ))
cp.show_plot()
```

```python
import MetaTrader5 as mt5
import pandas as pd
import plotly.graph_objects as go
import datetime as dt
from datetime import datetime
```

```python
mt5.initialize()
```

```python
timeframe=mt5.TIMEFRAME_H1
symbol = "EURUSD"
number_of_data = 600
PIP_LOCATION = 1e-5
```

```python
from_date = datetime.now()
# Extract n Ticks before now
rates = mt5.copy_rates_from(symbol, timeframe, from_date, number_of_data)
# Transform Tuple into a DataFrame
df_rates = pd.DataFrame(rates)
df_rates["time"] = pd.to_datetime(df_rates["time"], unit="s")
```

```python
df_rates.rename(columns={
    'close' : 'bid_c',
    'open' : 'bid_o',
    'high' : 'bid_h',
    'low' : 'bid_l',
}, inplace=True)
```

```python
ask_h_list = []
ask_l_list = []
ask_o_list = []
ask_c_list = []
spread_list = []
for i,t in enumerate(df_rates.time):
    ticks = mt5.copy_ticks_range(symbol,t,t+ dt.timedelta(hours=1),mt5.COPY_TICKS_ALL )
    df_ticks = pd.DataFrame(ticks)
#     print(df_ticks)
#     df_ticks["time"] = pd.to_datetime(df_ticks["time"], unit="s")
    df_ticks['spread'] = df_ticks.ask-df_ticks.bid
    ask_h = df_ticks.ask.max()
    ask_l = df_ticks.ask.min()
    ask_o = df_ticks.ask.iloc[0]
    ask_c = df_ticks.ask.iloc[-1]
    ask_h_list += [ask_h]
    ask_l_list += [ask_l]
    ask_o_list += [ask_o]
    ask_c_list += [ask_c]
    spread_list += [df_ticks['spread'].max()]
    
```

```python
df_rates['ask_h'] = ask_h_list
df_rates['ask_l'] = ask_l_list
df_rates['ask_o'] = ask_o_list
df_rates['ask_c'] = ask_c_list
df_rates.spread = spread_list

df_rates['mid_h'] = (df_rates.ask_h + df_rates.bid_h)/2
df_rates['mid_l'] = (df_rates.ask_l + df_rates.bid_l)/2
df_rates['mid_o'] = (df_rates.ask_o + df_rates.bid_o)/2
df_rates['mid_c'] = (df_rates.ask_c + df_rates.bid_c)/2
```
